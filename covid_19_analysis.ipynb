```python
%%bigquery
select cast(last_update as date) as date, sum(confirmed) as confirmed, sum(deaths) as deaths, sum(recovered) as recovered
from covid_19_modeled.Event_Beam e join covid_19_modeled.Location_Beam l
on e.location_id = l.id
where country = 'Mainland China' 
group by date
order by date desc
limit 12
```
